# 데이터 전처리

## 기간에서 연도를 분리하기

```python
# int("2014 1/4".split()[0]) # df["기간"]에서 연도만 추출
```

```python
df["연도"] = df["기간"].map(lambda x : int(x.split()[0])) 
# lambda : 익명함수 x라는 변수 받음
# lambda가 받을 함수 = 기간  / 기간을 x로 받음 / x를 split한게 -> 연도 칼럼으로 들어감
df.head()
```

```python
int("2014 1/4".split()[1].split("/")[0])
```
